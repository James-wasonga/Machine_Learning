```python
age = [10, 15, 20]
insurance_rate = [2, 3, 4]

for age, insurance_rate in zip(age, insurance_rate):
    insurance_per = age * insurance_rate

print(insurance_per)
```

```python
age = [10, 15, 20]
insurance_rate = [2, 3, 4]

insurance_per = [age * insurance_rate for age, insurance_rate in zip(age, insurance_rate)]
print(insurance_per)
```
